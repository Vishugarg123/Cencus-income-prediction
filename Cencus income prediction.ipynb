```python
x=x.drop(['Workclass', 'Race', 'Sex',
            'Capital_loss', 'Native_country'], axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
for col in x.columns:
        scaler = StandardScaler()
        x[col] = scaler.fit_transform(x[col].values.reshape(-1, 1))
    
```

```python
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=42)
```

```python
ros.fit(x,y)
```

```python
x_resampled,y_resampled=ros.fit_resample(x, y)
```

```python
round(y_resampled.value_counts(normalize=True) * 100, 2).astype('str') + ' %'
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_resampled,y_resampled,test_size=0.20,random_state=0)
```

```python
print("x_train shape:", x_train.shape)
print("x_test shape:", x_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
k_fold = KFold(n_splits=10, shuffle=True, random_state=0)
```

```python
clf = [LogisticRegression(),KNeighborsClassifier(n_neighbors = 13),DecisionTreeClassifier(),
       RandomForestClassifier(n_estimators=13),GaussianNB()
      ,XGBClassifier()]
def model_fit():
    scoring = 'accuracy'
    for i in range(len(clf)):
        score = cross_val_score(clf[i], x , y, cv=k_fold, n_jobs=1, scoring=scoring)
        print("Score of Model",clf[i],":",round(np.mean(score)*100,2))

model_fit()
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
RFC=RandomForestClassifier()
```

```python
n_estimators = [int(x) for x in np.linspace(start = 20, stop = 140, num = 4)]

max_features = ['auto', 'sqrt']

max_depth = [int(x) for x in np.linspace(40, 150, num = 15)]
max_depth.append(None)

min_samples_split = [ 5, 10]

min_samples_leaf = [ 2, 4]

bootstrap = [True, False]


grid = {"max_depth": max_depth,
              "max_features": max_features,
              "min_samples_split": min_samples_split,
              "min_samples_leaf": min_samples_leaf,
              "bootstrap": bootstrap,
              "n_estimators" :n_estimators,
              "criterion": ["gini"]}
```

```python
RSRFC=RandomizedSearchCV(RFC,grid,cv=k_fold,scoring="accuracy", n_jobs= 4)
RSRFC.fit(x_train,y_train)
RSRFC.best_params_
```

```python
xgb= XGBClassifier(random_state=0,min_child_weight=5,max_depth= 15,
                    learning_rate = 0.05,gamma= 0.1,
                    colsample_bytree= 0.4)
xgb.fit(x_train,y_train)
y_pred_xgb = xgb.predict(x_test)
```

```python
print('XG Boost Classifier:')
print('Accuracy score:', round(accuracy_score(y_test, y_pred_xgb) * 100, 2))
print('F1 score:', round(f1_score(y_test, y_pred_xgb) * 100, 2))
```

```python
from sklearn.metrics import roc_curve, auc
plt.figure(figsize = (15, 8))
plt.plot([0,1], [0,1], 'r--')

probs = xgb.predict_proba(x_test)
probs = probs[:, 1]
fpr, tpr, thresholds = roc_curve(y_test, probs)
roc_auc = auc(fpr, tpr)

label = 'XG boost classifier' + ' AUC:' + ' {0:.2f}'.format(roc_auc)
plt.plot(fpr, tpr, c = 'orange', label = label, linewidth = 2)
    
plt.xlabel('False Positive Rate', fontsize = 12)
plt.ylabel('True Positive Rate', fontsize = 12)
plt.title('Receiver Operating Characteristic', fontsize = 15)
plt.legend(loc = 'lower right',fontsize = 12)
plt.show()
```
